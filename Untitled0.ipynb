```python
import pandas as pd

melbourne_file_path = "https://raw.githubusercontent.com/erkansirin78/datasets/master/Melbourne_housing_FULL.csv"

df = pd.read_csv(melbourne_file_path)

features = ['Rooms', 'Distance', 'Bathroom', 'Car']
target = 'Price'

df_clean = df[features + [target]].dropna()


X_raw = df_clean[features].values.tolist()
y = df_clean[target].values.tolist()

print("Dataset successfully loaded and filtered!")
print(f"Total clean property rows ready to use: {len(X_raw)}")
print(f"Example data row (Rooms, Distance, Bathroom, Car): {X_raw[0]}")
print(f"Example target row (Price): ${y[0]:,.2f}")
```

```python

m = len(X_raw)
n_features = len(X_raw[0])

#calculating mean
means = [0.0] * n_features
for j in range(n_features):
    column_sum = sum(X_raw[i][j] for i in range(m))
    means[j] = column_sum / m

#calculation standard deviation
stdevs = [0.0] * n_features
for j in range(n_features):
    variance_sum = sum((X_raw[i][j] - means[j]) ** 2 for i in range(m))
    stdevs[j] = (variance_sum / m) ** 0.5
    if stdevs[j] == 0:
        stdevs[j] = 1.0
```

```python
import time

# PURE PYTHON IMPLEMENTATION

X = X_raw

alpha = 0.01
iterations = 1000

m = len(X)
n = len(X[0])

weights = [0.0] * n
cost_history = []

print("Starting training via Pure Python loops...")
start_time = time.time()

for it in range(iterations):

    predictions = []

    for i in range(m):
        pred = sum(X[i][j] * weights[j] for j in range(n))
        predictions.append(pred)


    cost_sum = sum((predictions[i] - y[i]) ** 2 for i in range(m))
    cost = cost_sum / (2 * m)
    cost_history.append(cost)


    if (it + 1) % 100 == 0:
        print(f"Iteration {it + 1}/{iterations} | Cost: {cost:,.2f}")


    gradients = [0.0] * n

    for j in range(n):
        grad_sum = sum((predictions[i] - y[i]) * X[i][j] for i in range(m))
        gradients[j] = grad_sum / m


    for j in range(n):
        weights[j] -= alpha * gradients[j]

end_time = time.time()
pure_python_time = end_time - start_time

print("\n--- Training Finished! ---")
print(f"Time Taken: {pure_python_time:.4f} seconds")
print(f"Final learned weights: {weights}")


y_pred = []

for i in range(m):
    pred = sum(X[i][j] * weights[j] for j in range(n))
    y_pred.append(pred)



y_true = y


mae = sum(abs(y_true[i] - y_pred[i]) for i in range(m)) / m


mse = sum((y_true[i] - y_pred[i]) ** 2 for i in range(m)) / m


rmse = mse ** 0.5


y_mean = sum(y_true) / m

ss_res = sum((y_true[i] - y_pred[i]) ** 2 for i in range(m))
ss_tot = sum((y_true[i] - y_mean) ** 2 for i in range(m))

r2_score = 1 - (ss_res / ss_tot)
print("\n--- Performance Metrics ---")
print(f"MAE  = {mae:.4f}")
print(f"RMSE = {rmse:.4f}")
print(f"R²   = {r2_score:.4f}")
```

```python
import numpy as np
import time

# NUMPY IMPLEMENTATION


X_np = np.array(X_raw, dtype=np.float64)
y_np = np.array(y, dtype=np.float64).reshape(-1, 1)


mean_np = np.mean(X_np, axis=0)
std_np = np.std(X_np, axis=0)

X_scaled_np = (X_np - mean_np) / np.where(std_np == 0, 1.0, std_np)


X_bias_np = np.hstack((np.ones((X_scaled_np.shape[0], 1)), X_scaled_np))


m_np, n_np = X_bias_np.shape


alpha = 0.01
iterations = 1000


weights_np = np.zeros((n_np, 1))


numpy_cost_history = []

print("Starting training via NumPy vectorized operations...")
start_time = time.time()

for it in range(iterations):


    predictions = X_bias_np @ weights_np


    errors = predictions - y_np


    cost = (1 / (2 * m_np)) * np.sum(errors ** 2)
    numpy_cost_history.append(cost)

    gradients = (1 / m_np) * (X_bias_np.T @ errors)


    weights_np -= alpha * gradients

end_time = time.time()
numpy_time = end_time - start_time

print("\n--- NumPy Training Finished! ---")
print(f"Time Taken: {numpy_time:.4f} seconds")
print(f"Final weights found: {weights_np.flatten().tolist()}")
print(f"Speed boost factor: {pure_python_time / numpy_time:.1f}x faster!")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
import time

# SCIKIT-LEARN IMPLEMENTATION


scaler = StandardScaler()
X_scaled_sku = scaler.fit_transform(X_raw)


model = LinearRegression()

print("Fitting model via Scikit-learn...")
start_time = time.time()

model.fit(X_scaled_sku, y)

sklearn_time = time.time() - start_time


sklearn_weights = [model.intercept_] + model.coef_.tolist()

print("\n--- Scikit-learn Fitting Finished! ---")
print(f"Time Taken: {sklearn_time:.4f} seconds")
print(f"Final weights found: {sklearn_weights}")


sklearn_predictions = model.predict(X_scaled_sku)


y_true = y
y_pred = sklearn_predictions

m = len(y_true)


mae_sklearn = sum(abs(y_true[i] - y_pred[i]) for i in range(m)) / m


mse_sklearn = sum((y_true[i] - y_pred[i]) ** 2 for i in range(m)) / m

rmse_sklearn = mse_sklearn ** 0.5


y_mean = sum(y_true) / m

ss_res = sum((y_true[i] - y_pred[i]) ** 2 for i in range(m))
ss_tot = sum((y_true[i] - y_mean) ** 2 for i in range(m))

r2_sklearn = 1 - (ss_res / ss_tot)

print("\n--- Scikit-learn Performance Metrics ---")
print(f"MAE  = {mae_sklearn:.4f}")
print(f"RMSE = {rmse_sklearn:.4f}")
print(f"R²   = {r2_sklearn:.4f}")
```
